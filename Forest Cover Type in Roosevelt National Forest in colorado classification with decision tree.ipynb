### Forest Cover Type in Roosevelt National Forest in colorado

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('covtype.csv')
```

```python
df
```

```python
df['Cover_Type'].value_counts()
```

```python
import sklearn.model_selection as ml
import sklearn.tree as tree
import sklearn.metrics as metrices
```

#### Train test split

```python
y=df.Cover_Type
x=df.drop('Cover_Type',axis=1)
```

```python
print(y.shape,x.shape)
```

```python
x_train,x_test,y_train,y_test=ml.train_test_split(x,y,random_state=1)
```

#### Model building

```python
x_train
```

```python
x_train.info()
```

```python
x_train['Soil_Type1'].value_counts()
```

```python
x_train['Wilderness_Area4'].value_counts()
```

```python
x_train.describe()
```

```python
for i in range(1,40):
    for j in range(i,41):
        if sum(x_train['Soil_Type'+str(i)]*x_train['Soil_Type'+str(j)])==0:
            pass
        else:
            print(sum(x_train['Soil_Type'+str(i)]*x_train['Soil_Type'+str(j)]))
```

```python
xtrain,xval,ytrain,yval=ml.train_test_split(x_train,y_train)
```

```python
algo1=tree.DecisionTreeClassifier()
algo1.fit(xtrain,ytrain)
```

```python
algo1.score(xtrain,ytrain)
```

```python
algo1.score(xval,yval)
```

```python
algo1.get_depth()
```

```python
a=np.arange(35,43)
testscore=np.zeros(len(a))
trainscore=np.zeros(len(a))
for i in range(len(a)):
    algo1=tree.DecisionTreeClassifier(max_depth=a[i])
    algo1.fit(xtrain,ytrain)
    trainscore[i]=algo1.score(xtrain,ytrain)
    testscore[i]=algo1.score(xval,yval)
print(trainscore,testscore)
```

Best tree is with max_depth=38

```python
algo1=tree.DecisionTreeClassifier(max_depth=38)
algo1.fit(xtrain,ytrain)
algo1.score(xval,yval)
```

### Results

```python
y_pred=algo1.predict(x_test)
```

```python
from sklearn import metrics as mts
mts.confusion_matrix(y_test,y_pred)
```

```python
report=mts.classification_report(y_test,y_pred)
print(report)
```
