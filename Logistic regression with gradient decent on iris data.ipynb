### Gradient decent for logistic regression

```python
def grad_point(x,y,m,C):
    gx=(m*x).sum()
    hx=1/(1+np.exp(-gx))
    return C*(hx-y)*x+2*m
```

```python
def grad(x,y,m,C,batch):
    [d0,d1]=x.shape
    grad=np.zeros(d1)
    for i in random.sample(range(d0),batch):
        grad+=grad_point(x[i,:],y[i],m,C)
    return grad/batch
```

```python
def logistic_regression(x,y,lr,C,max_iter,batch):
    [d0,d1]=x.shape
    m=np.zeros(d1)
    for i in range(max_iter):
        m-=lr*grad(x,y,m,C,batch)
    return m
```

```python
def predict(x,m):
    [d0,d1]=x.shape
    y=np.zeros(d0)
    for i in range(d0):
        gx=(x[i,:]*m).sum()
        y[i]=1 if 1/(1+np.exp(-gx))>0.5 else 0
    return y
```

```python
[d0,d1]=x.shape
[dt0,dt1]=x_test.shape
```

### Implementation

```python
m=logistic_regression(x,y,.01,1,10,70)
print(m)
```

```python
y_predict=predict(x_test,m)
```
