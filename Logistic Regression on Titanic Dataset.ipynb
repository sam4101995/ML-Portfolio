## Logistic Regression on Titanic dataset

```python
import numpy as np
from sklearn import linear_model as lm
from sklearn import metrics
import pandas as pd
#from sklearn import datasets
```

```python
df=pd.read_csv('Downloads/titanic_x_y_train.csv')
df_test=pd.read_csv('Downloads/titanic_x_test.csv')
df_test
```

### Droping features with missing values

```python
df1=df.drop(['Name','Ticket','Cabin','Age','Embarked'],axis=1)
df_test=df_test.drop(['Name','Ticket','Cabin','Age','Embarked'],axis=1)
df1
```

```python
df2=df1.dropna()
df2.shape
df_test.info
```

```python
df2['Sex']=df2['Sex'].map({'female':-1,'male':1})
df_test['Sex']=df_test['Sex'].map({'female':-1,'male':1})
y=df2['Survived'].values
df2=df2.drop(['Survived'],axis=1)
df2
```

### Composite Feature creation

```python
d1=df2.shape[1]
for i in range(d1):
    for j in range(i,d1):
        if (df2.iloc[:,i]*df2.iloc[:,j]).var()==0:
            pass
        else:
            df2[str(i)+str(j)]=df2.iloc[:,i]*df2.iloc[:,j]

d2=df2.shape[1]
for i in range(d1):
    for j in range(i,d2):
        if (df2.iloc[:,i]*df2.iloc[:,j]).var()==0:
            pass
        else:
            df2[str(i)+str(j)]=df2.iloc[:,i]*df2.iloc[:,j]
d3=df2.shape[1]
for i in range(d1):
    for j in range(i,d3):
        if (df2.iloc[:,i]*df2.iloc[:,j]).var()==0:
            pass
        else:
            df2[str(i)+str(j)]=df2.iloc[:,i]*df2.iloc[:,j]
df2.shape
```

```python
for i in range(d1):
    for j in range(i,d1):
        df_test[str(i)+str(j)]=df_test.iloc[:,i]*df_test.iloc[:,j]
for i in range(d1):
    for j in range(i,d2):
        df_test[str(i)+str(j)]=df_test.iloc[:,i]*df_test.iloc[:,j]
for i in range(d1):
    for j in range(i,d3):
        df_test[str(i)+str(j)]=df_test.iloc[:,i]*df_test.iloc[:,j]
df_test.shape 
```

```python
df_test=df_test[df2.columns]
df_test.shape
```

```python
df2
```

```python
x=df2.values
x_test=df_test.values
```

```python
print(x.shape)
print(y.shape)
```

### Standardization

```python
norm=np.zeros([2,x.shape[1]])
for i in range(x.shape[1]):
    norm[0,i]=x[:,i].mean()
    norm[1,i]=np.sqrt(x[:,i].var())
    x[:,i]=(x[:,i]-norm[0,i])/norm[1,i]
x
```

```python
for i in range(x_test.shape[1]):
    x_test[:,i]=(x_test[:,i]-norm[0,i])/norm[1,i]
```

```python
y
```

### Logistic Regression

```python
algo=lm.LogisticRegression(solver='saga',max_iter=2000)
algo.fit(x,y)
#y_pred=algo.predict(x)
#algo.score(y,y_pred)
#mat=metrics.confusion_matrix(y,y_pred)
#mat
```

```python
y_pred=algo.predict(x_test)
```

```python
y_pred
```
